### Introduction
The aim of this project is to scrape an e-commerce website using selenium and obtain information of all available smart watches on the first page.


#### Imports

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
```

#### Initialization

```python
# initialize driver
driver=webdriver.Chrome('chromedriver')
#open website
url='https://www.jumia.co.ke/'
driver.get(url)
#maximize window
driver.maximize_window()
time.sleep(1)
# close the popup
popup=driver.find_element(By.XPATH, '//div[@id="pop"]/div/section/button')
popup.click()
time.sleep(1)
# search for smart watch
search=driver.find_element(By.XPATH,'//input[@name="q"]')
search.click()
time.sleep(1)
search.send_keys('smart watch')
search.send_keys(Keys.ENTER)
```

#### Target data

```python
# create lists of items to be scraped
product_name=[]
product_review=[]
product_price=[]
product_oldprice=[]
product_discount=[]
#loop through the 48 items on the first page
for i in range(48):
    try:
        product_name.append(driver.find_elements(By.XPATH, '//h3[@class="name"]')[i].text)
    except:
        product_name.append('')
    try:
        product_review.append(driver.find_elements(By.XPATH, '//div[@class="stars _s"]')[i].text)
    except:
        product_review.append('')
    try:
        product_price.append(driver.find_elements(By.XPATH, '//div[@class="prc"]')[i].text)
    except:
        product_price.append('')
    try:
        product_oldprice.append(driver.find_elements(By.XPATH, '//div[@class="old"]')[i].text)
    except:
        product_oldprice.append('')
    try:
        product_discount.append(driver.find_elements(By.XPATH, '//div[@class="bdg _dsct _sm"]')[i].text)
    except:
        product_discount.append('')
    #create a pandas dataframe and store the scraped items
    watch_df=pd.DataFrame({'Name':product_name,
                     'Review':product_review,
                     'Price':product_price,
                     'Old-price':product_oldprice,
                     'Discount':product_discount
    })
driver.quit()
```
